## tsf 파일 처리법

```python
import pandas as pd
from aeon.datasets import load_from_tsf_file #numpy 버전을 aeon에 맞춰 설치하여야 함: 2.0.0 이하
import matplotlib.pyplot as plt

df = load_from_tsf_file('data/australian_electricity_demand_dataset.tsf')
df
```

```python
ts = df[0]['series_value'][0]

plt.plot(pd.date_range(df[0]['start_timestamp'][0], periods=len(ts), freq='0.5h'), ts)
plt.show()
```
